```python
import pandas as pd
import numpy as np
import math

import matplotlib.pyplot as plt

import statsmodels.api as sm # OLS
from sklearn.linear_model import Lasso #LASSO
from scipy import stats

from itertools import islice
from datetime import datetime

import pandas_market_calendars as mcal
```

```python
def power_utility(x,gamma):
    u=(x**(1-gamma)-1)/(1-gamma)
    return u

def log_utility(x):
    u=np.log(x)
    return u

def exp_utility(x,gamma):
    u=-np.exp(-gamma*x)
    return u
```

```python

x = np.linspace(0.01, 5, 400)

for gamma in [0.1,0.2,0.5,1]:
    if gamma != 1:
        y = power_utility(x,gamma)
    else:
        y = log_utility(x)
    plt.plot(x,y,label=f"$\gamma$={gamma}")


plt.xlabel('$x$')
plt.ylabel('utility')
plt.legend()

#exp = exp_utility(x,gamma)
```
